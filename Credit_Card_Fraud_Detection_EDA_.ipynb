## 10. Check the distribution of columns. List down columns that are normally distributed. List down columns that are not normally distributed.

```python
from scipy import stats
normally_distributed_columns = []

# List of columns that do not appear to be normally distributed
not_normally_distributed_columns = []

# Set a significance level for the normality test (e.g., 0.05)
alpha = 0.05

# Iterate through columns in the DataFrame
for column in df.columns:    
    # Perform a normality test (Shapiro-Wilk) on the data
    p_value = stats.shapiro(df[column])[1]
    
    # Check if the column is normally distributed based on the p-value
    if p_value > alpha:
        normally_distributed_columns.append(column)
    else:
        not_normally_distributed_columns.append(column)

# List down columns that are normally distributed
print("Columns that are normally distributed:")
print(normally_distributed_columns)

# List down columns that are not normally distributed
print("Columns that are not normally distributed:")
print(not_normally_distributed_columns)
```

## 18. With the help of hypothesis testing check whether fraudulent transactions of higher value than normal transactions?

```python
from scipy import stats
fraudulent = df[df['Class'] == 1]['Amount']
non_fraudulent = df[df['Class'] == 0]['Amount']

# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(fraudulent, non_fraudulent, equal_var=False)

# Define the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to make a decision
if p_value < alpha:
    print("Reject the null hypothesis: Fraudulent transactions have a significantly higher mean amount.")
else:
    print("Fail to reject the null hypothesis: No significant difference in mean amount between fraudulent and non-fraudulent transactions.")
```

```python
# 21.11 How many Boolean columns are there in the data set and out of those howmany
# are imbalanced?
import pandas as pd

# Assuming df is your DataFrame

# Identify Boolean columns
boolean_columns = df.select_dtypes(include=bool).columns

# Count the number of Boolean columns
num_boolean_columns = len(boolean_columns)

# Initialize a counter for imbalanced columns
num_imbalanced_columns = 0

# Check for imbalance in each Boolean column
for col in boolean_columns:
    value_counts = df[col].value_counts()
    if len(value_counts) == 2:
        min_count = min(value_counts)
        max_count = max(value_counts)
        imbalance_ratio = min_count / max_count
        if imbalance_ratio < 0.1 or imbalance_ratio > 0.9:
            num_imbalanced_columns += 1

print("Number of Boolean Columns:", num_boolean_columns)
print("Number of Imbalanced Boolean Columns:", num_imbalanced_columns)
```

```python
# 21.12 Check for duplicate records across all columns (Check Warning Section
import pandas as pd

# Assuming df is your DataFrame

# Chunk size for processing the data
chunk_size = 1000

# Set to store duplicate records
duplicate_records = set()

# Iterate over the DataFrame in chunks
for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i+chunk_size]
    duplicates_in_chunk = chunk.duplicated()
    # Add duplicate records to the set
    duplicate_records.update(chunk[duplicates_in_chunk].index)

# Extract the duplicate records from the DataFrame
duplicate_records_df = df.loc[list(duplicate_records)]

# Display the duplicate records
print("Duplicate Records:")
print(duplicate_records_df)
```

```python
# 21.13 Is there any imbalance in the categorical columns? (for example Gender Male and
# Female in which Male is 95% and Female is just 5%- How many columns are
# categorical?)
  
import pandas as pd

# Assuming df is your DataFrame

# Identify categorical columns
categorical_columns = df.select_dtypes(include='object').columns

# Count the number of categorical columns
num_categorical_columns = len(categorical_columns)

# Initialize a counter for imbalanced columns
num_imbalanced_categorical_columns = 0

# Check for imbalance in each categorical column
for col in categorical_columns:
    value_counts = df[col].value_counts(normalize=True)
    if len(value_counts) == 2:
        min_ratio = min(value_counts)
        max_ratio = max(value_counts)
        if min_ratio < 0.1 or max_ratio < 0.1:
            num_imbalanced_categorical_columns += 1

print("Number of Categorical Columns:", num_categorical_columns)
print("Number of Imbalanced Categorical Columns:", num_imbalanced_categorical_columns)
```

```python
# 21.14.Are those categories in sync with the domain categories?
# Define predefined domain categories
domain_categories = ["Category1", "Category2", "Category3", "Category4", "Category5"]

# Get the unique values in each of the specified columns
column_values = {}
for column in df.columns:
    column_values[column] = df[column].unique()

# Function to check if categories are in sync
def are_categories_in_sync(domain_cats, user_cats):
    # Convert lists to sets for efficient comparison
    domain_set = set(domain_cats)
    user_set = set(user_cats)

    # Check if the sets are equal (categories match)
    if domain_set == user_set:
        return True
    else:
        return False

# Check if user categories are in sync with domain categories for each column
for column, values in column_values.items():
    sync_status = are_categories_in_sync(domain_categories, values)
    if sync_status:
        print(f"{column}: User categories are in sync with domain categories.")
    else:
        print(f"{column}: User categories do not match domain categories.")
```

```python
#  Feature Engineering (Feature Selection and Feature Creation)
# To define the userdefined function
def AmountCat(data):
    if data>0 and data<100:
        return 'Low'
    if data>=100 and data<500 :
        return "Medium"
    if data>=500 and data<1000 :
        return "High"
    if data>=1000:
        return "Very High"
df["AmountCat"] =df['Amount'].apply(AmountCat)
```
